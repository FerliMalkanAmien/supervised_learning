```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,10,100)
y = np.sqrt(X)
y1 = np.sin(X)
y2 = np.cos(X)
```

# Inisalisasi Figure dan Axes

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(X, y)
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(X, y1)
ax.plot(X, y2)
```

### shortcut

```python
plt.figure(figsize=(6,6))
plt.plot(X, y1, 'r-', linewidth=0.5) # r untuk warna, dan -. untuk syle garis
plt.plot(X, y2, 'bo', markersize=3) # o untuk bulat, v segitiga terbalik
plt.xlim(0, 10)
plt.ylim(-2, 2)
plt.axis('equal') # rasio x dan y jadi setara
plt.title('Sin dan Cos')
plt.xlabel('X')
plt.ylabel('y')
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()

ax.plot(X, y1, 'r-', linewidth=0.5) # r untuk warna, dan -. untuk syle garis
ax.plot(X, y2, 'bo', markersize=3) # o untuk bulat, v segitiga terbalik
ax.set_xlim(0, 10)
ax.set_ylim(-2, 2)
ax.axis('equal')# rasio x dan y jadi setara
ax.set_title('Sin dan Cos', fontsize=14)
ax.set_xlabel('X', fontsize=10)
ax.set_ylabel('y', fontsize=10)
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()

ax.plot(X, y1, 'r-', linewidth=0.5) # r untuk warna, dan -. untuk syle garis
ax.plot(X, y2, 'bo', markersize=3) # o untuk bulat, v segitiga terbalik
ax.set(xlim=(0, 10), ylim=(-2, 2), title = "Sin dan Cos", xlabel='X', ylabel='y')
ax.axis('equal'); # rasio x dan y jadi setara
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()

ax.plot(X, y1, 'r-', linewidth=0.5, label='Sin') # r untuk warna, dan -. untuk syle garis
ax.plot(X, y2, 'bo', markersize=3, label='Cos') # o untuk bulat, v segitiga terbalik
ax.set(xlim=(0, 10), ylim=(-2, 2), title = "Sin dan Cos", xlabel='X', ylabel='y')
ax.axis('equal'); # rasio x dan y jadi setara
ax.legend(loc='lower center', ncol=2, shadow=True, borderpad=1)
```
